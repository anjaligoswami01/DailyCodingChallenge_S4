```python
!pip install sentence-transformers

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
sentences = [
    "I love playing football",
    "I enjoy playing soccer",
    "The sky is clear today",
    "It is a sunny day outside",
    "I am learning machine learning"
]
```

```python
embeddings = model.encode(sentences)
```

```python
similarity_matrix = cosine_similarity(embeddings)

df = pd.DataFrame(similarity_matrix, index=sentences, columns=sentences)
df
```

```python
def compare_sentences(s1, s2):
    emb1 = model.encode([s1])
    emb2 = model.encode([s2])
    score = cosine_similarity(emb1, emb2)[0][0]
    return score

print("Football vs Soccer:", compare_sentences(sentences[0], sentences[1]))
print("Sky vs Sunny:", compare_sentences(sentences[2], sentences[3]))
print("Football vs ML:", compare_sentences(sentences[0], sentences[4]))
```
